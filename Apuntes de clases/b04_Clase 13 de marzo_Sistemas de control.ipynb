```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

$$\dot{x}+c{x}=0$$
$$V(x)=\frac{1}{2}{x^2}$$


```python
def f(x,t):
    dx=-x
    return dx

#t = np.linspace(0,20,200)
t=np.arange(0,3,.1)
x0=-.0043
ys=odeint(f,x0,t)

plt.plot(t,ys[:,0])
```

$$\dot{x_1}=-x_1$$
$$\dot{x_2}=-2x_2$$

$$V(x)=\frac{1}{2}{x^2}$$
$$\dot{V}(x) = {x}\dot{x} = [x_1,x_2][{\dot{x}_1}{\dot{x}_2}]^T=x_1\dot{x}_1+x_2\dot{x}_2$$

$$\dot{V}(x_1,x_2)=-{x_1}^2-2{x_2}^2$$

```python
def f(x,t):
    x1,x2=x
    dx1=-x1
    dx2=-2*x2
    return [dx1,dx2]

#t=np.linspace(0,20,200)
t=np.arange(0,3,.1)
x0=[4,4]
ys=odeint(f,x0,t)
```

```python
y1=np.linspace(-8.0,8.0,20)
y2=np.linspace(-8.0,8.0,20)
X,Y=np.meshgrid(y1,y2)

U,V=f([X,Y],0)
plt.figure(figsize=(9,8))
Q=plt.quiver(X,Y,U,V,color='r')
plt.plot(ys[:,0],ys[:,1],'b-') #path
plt.plot([ys[0,0]], [ys[0,1]], 'o')#start
plt.plot([ys[-1,0]], [ys[-1,1]], 's')#end
```

```python
def f(x,t):
    u=-2*x**2
    dx1=x**2+u
    return dx1

#t=np.linspace(0,20,200)
t=np.arange(0,3,.1)
x0=[2,1]
ys=odeint(f,x0,t)

plt.plot(t,ys[:,0])
plt.plot(t,ys[:,1])
```

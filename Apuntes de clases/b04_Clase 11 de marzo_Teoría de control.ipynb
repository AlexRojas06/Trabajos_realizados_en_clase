```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def sys(X,t):
    x1, x2 = X
    m = 1
    b = 1
    R = 1
    g = 9.8
    dx1 = x2
    dx2 = -(1/(m*R**2))*x2 - (g/R)*(np.sin(x1))
    return [dx1,dx2]

x0 = [0.1,0.1]
t = np.linspace(3,20,400)
sol = odeint(sys, x0, t)

plt.plot(t, sol[:,0], 'b', label = 'x1(t)')
plt.plot(t, sol[:,1], 'g', label = 'x2(t)')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

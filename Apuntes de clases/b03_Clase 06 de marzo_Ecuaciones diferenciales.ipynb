```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r=1
def poblacion(x,t):
    return r*x*(1-x)
```

```python
x0=0.05
tt=np.linspace(0,10)
xx=odeint(poblacion,x0,tt)
plt.plot(tt,xx)
plt.xlabel('$t$', fontsize=18)
plt.ylabel('$x$', fontsize=18)
plt.show()
```

```python
def logi_sol(t,x0,r):
    return 1/(1+(1/x0-1)*np.exp(-r*t))
```

```python
t=np.linspace(0,10)
for r in np.arange(-1,1,.1):
    plt.plot(t,logi_sol(t,0.05,r))
plt.xlabel('$t$',fontsize=18)
plt.xlabel('$t$',fontsize=18)
plt.show()
```

```python
def mapa_logistico(r,x):
    return r*x*(1-x)
```

```python
r=2
#si cambias la r, cambia el comportamiento
r=2.5
r=3.1
r=3.6
n=100
x=np.zeros(n)
x[0]=0.3

for i in range (n-1):
    x[i+1]=mapa_logistico(r,x[i])

plt.plot(np.arange(0,n),x)
```

```python
r=2
n=400
r=np.linspace(.5,4.0,n)
x=1e-5*np.ones(n)

for i in range (100):
    x=mapa_logistico(r,x)
    if i>80:
        plt.plot(r,x,",r",alpha=.2)
```
